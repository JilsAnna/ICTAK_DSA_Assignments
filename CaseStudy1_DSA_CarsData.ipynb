# Read the dataset to the python environment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("E:\ICTAK-DSA\cars_data.csv")
df=pd.DataFrame(data)
df
```

# Check for the null values present in the dataset

```python
df.isnull().sum()
```

# Plot a bar graph of male vs female buyers participated in the sales

```python
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(10,7))
sales_partpn=sns.countplot(df['Buyer Gender'])
plt.xlabel("Gender")
plt.ylabel("Count of Buyers")
plt.grid(axis = "y")
plt.title("Count of Male v/s Female Buyers Participated in the sales",fontsize=13)
```

# Find the top 5 cars based on their sales price

```python
topcar=df.nlargest(5,'Sale Price')
topcar.loc[:,[ "Model", "Sale Price"]]
```

# Find the least 5 cars based on their Resell price

```python
leastcar=df.nsmallest(5,'Resell Price')
leastcar.loc[:,[ "Model", "Resell Price"]]
```
