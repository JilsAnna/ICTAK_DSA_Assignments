# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# Pairplot

```python
sns.pairplot(df)
```

# HeatMap

```python
sns.heatmap(df.corr(), annot=True)
plt.title('Heatmap of Corelation of Iris Data', fontsize=12)
```

# ViolinPlot

```python
sns.violinplot(x='PL', y='Classification', data=df)
plt.title('Violinplot of PL', fontsize=12)
```

# Box Plot

```python
sns.boxplot(x='PL', y='Classification', data=df)
plt.title('Boxplot of PL', fontsize=12)
```

# ScatterPlot

```python
sns.scatterplot(data=df, x="SL", y="SW", hue="Classification")
plt.title("Sepal Length and Width of Different Classification",fontsize="14")
```

```python
sns.scatterplot(data=df, x="PL", y="PW", hue="Classification")
plt.title("Petal Length and Width of Different Classification",fontsize="14")
```
